#### Adding Conditions

```python
def generate_data(n, generator_type):
    _values = []
    
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        _values.append(e)
    
    return _values
data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```

```python
def generate_data(n, generator_type):
    _values = []

    for i in range(n):
        e = generator_type()
        _values.append(e)

    return _values

data = generate_data(100, np.random.uniform)
plt.plot(data)
plt.show()
```

### Exercises

```python
def factorial(n):
    k = 1
    for i in range(n):
        k = k * (i + 1)
    
    return k

factorial(4)
```

```python
def factorial(n,f = lambda x: x):
    k = 1

    for i in range(f(n)):
        k = k * (i + 1)
    return k

factorial(9) # default
```

```python
f = lambda x: x**2 + 1 if x % 2 == 0 else x**2
factorial(3, f) # odd (equivalent to factorial(9))
```

```python
factorial(2, f) # even (equivalent to factorial(5))
```

Solution to Exercise

```python
from numpy.random import uniform

def binomial_rv(n, p):
    count = 0

    for i in range(n):
        U = uniform()
        if U < p:
            count = count + 1 # Or count += 1
    return count

binomial_rv(10, 0.5)
```

```python
from numpy.random import uniform

def draw(k): # pays if k consecutive successes in a sequence
    payoff = 0
    count = 0
    
    for i in range(10):
        U = uniform()
        count = count + 1 if U < 0.5 else 0
        print(count) # print counts for clarity
        if count == k:
            payoff = 1
    
    return payoff

draw(3)
```

```python
def draw_new(k): # pays if k successes in a sequence
    payoff = 0
    count = 0

    for i in range(10):
        U = uniform()
        count = count + ( 1 if U < 0.5 else 0 )
        print(count)
        
        if count == k:
            payoff = 1

    return payoff

draw_new(3)
```

### Advanced Exercises

```python
def x(t):
    if t == 0:
        return 0
    
    if t == 1:
        return 1
    
    else:
        return x(t-1) + x(t-2)
```

```python
print([x(i) for i in range(10)])
```

```python
def recursion_factorial(n):
    if n == 1:
        return n
    
    else:
        return n * recursion_factorial(n-1)
```

```python
def recursion_factorial_simplified(n):
    return n * recursion_factorial(n-1) if n != 1 else n
```

```python
print([recursion_factorial(i) for i in range(1, 10)])
```

```python
print([recursion_factorial_simplified(i) for i in range(1, 10)])
```
